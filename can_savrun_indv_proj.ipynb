```python
rushhourmorning=df.loc[df["time_of_day"] =='rush_hour_morning']
rushhourmorninga=np.array(rushhourmorning['distance'])
afternoon=df.loc[df["time_of_day"] =='afternoon' ]
afternoonarray=np.array(afternoon['distance'])
rushhourevening=df.loc[df["time_of_day"] =='rush_hour_evening']
rushhoureveninga=np.array(rushhourevening['distance'])
latenight=df.loc[df["time_of_day"] =='late_night']
latenighta=np.array(latenight['distance'])
evening=df.loc[df["time_of_day"] == 'evening']  
eveninga=np.array(evening['distance'])
```

```python
passen1 =df[df["passenger_count"]==1]['distance']
passenmore1=df[df["passenger_count"]>1]['distance']
```

```python
fig, ax = plt.subplots(1, 3, figsize=(14,6))
passen1.plot(kind="hist",ax=ax[0],bins=40,color="c")
ax[0].set_title("Passenger number one")
passenmore1.plot(kind="hist", ax=ax[1], bins=40, color="m")
ax[1].set_title("Passenger number more than one")
sns.kdeplot(passen1, shade=True, label="1", ax=ax[2], color="c")
sns.kdeplot(passenmore1, shade=True, label="more than one", ax=ax[2], color="m")
ax[2].set_title("Comparison")
plt.suptitle("Distance Distributions")
plt.show()
```

```python
stats.ttest_ind(passen1,passenmore1,equal_var=False)
```

If we set the significance level as 0.05, we can reject the null hypothesis. So we can say that passenger group size has eï¬ect on the distance.

Null hypothesis: passenger group size has no eï¬ect on the distance 

Alternative Hypothesis : passenger group size has eï¬ect on the distance
